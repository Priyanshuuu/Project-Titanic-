```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

df=pd.read_csv(r'C:\Users\PRIYANSHU PANDEY\Downloads\titanic_train.csv')

t=df.copy()

Ns=t[t.Survived==0]

Ys=t[t.Survived==1]

Ns.Age.fillna(Ns.Age.mean(),inplace=True)
Ys.Age.fillna(Ys.Age.mean(),inplace=True)

newt=pd.concat([Ns,Ys])

del newt['Name']
del newt['Ticket']
del newt['Fare']
del newt['Cabin']
del newt['Embarked']

def f(s):
    if s=='male':
        return 1
    else :
        return 0

newt['Gender']=newt.Sex.apply(f)
del newt['Sex']

newt.head(30)
newt.describe()

female=newt[newt.Gender==0]
fems=female[female.Survived==1]
femns=female[female.Survived==0]

male=newt[newt.Gender==1]
ms=male[male.Survived==1]
mns=male[male.Survived==0]

sizes=[len(ms),len(fems),len(mns),len(femns)]

plt.axis("equal")                                

labels=['Males Survived','Females Survived','Males Not Survived','Females Not Survived']         

colors=['green','yellow','green','yellow']

explode=[0.05,0.05,0.05,0.05]                             

plt.title("Titanic Survive M/F",fontsize=35)             

plt.pie(sizes,colors=colors,labels=labels,explode=explode,autopct='%.1f%%',startangle=50,counterclock=False)

plt.show()


```
